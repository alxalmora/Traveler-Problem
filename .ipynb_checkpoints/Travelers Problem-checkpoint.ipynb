```python
import pandas as pd
import numpy as np
import mlrose
import json
import time
from matplotlib import pyplot as plt
print ('Librerías Cargadas.')
data = pd.read_csv('DataSet/data.csv', header = None, names=["id", "x", "y"])
print('CSV Cargado.')
```

```python
matrix = np.zeros(shape=(980,2))
print('Creando visualización de ciudades de luxemburgo')
for i in range(0,980):
    matrix[i][0] = data['x'][i]
    matrix[i][1] = data['y'][i]
plt.figure(figsize=(40,40));
plt.plot([p[0] for p in matrix], [p[1] for p in matrix], 'o');
plt.axis('scaled');
plt.axis("off");
plt.show();
```

```python
print("Creando Arreglo de localizaciones")
locations = [ (data['x'][i], data['y'][i]) for i in range(0,980)];
```

```python
fitness_coords = mlrose.TravellingSales(coords = locations)
```

```python
problem_fit = mlrose.TSPOpt(length = 980, fitness_fn = fitness_coords,maximize=False)
```

```python
arr= []
for mut_prob in range(0.1,1,0.1):
    for max_at in range(10,200,10):
        start_time = time.time()
        best_state, best_fitness = mlrose.genetic_alg(problem_fit, random_state = 149634, mutation_prob = 0.2, max_attempts = 100)
        elapsed_time = time.time() - start_time
        data = {"mutation_rate": mut_prob, "max_attempts":max_at, "best_fitness":best_fitness, "best_route":best_state, "time": time}
        arr.append(data)
```
